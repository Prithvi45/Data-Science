```python
import numpy as np
import matplotlib.pyplot as pp
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,40)
```

```python
x
```

```python
sinx = np.sin(x)
```

```python
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
```

```python
pp.plot(x,sinx)
pp.plot(x,cosx)

pp.legend(['sin(x)','cos(x)'])
```

```python
y = sinx * cosx
z = cosx**2 - sinx**2

pp.plot(x,y)
pp.plot(x,z)
```

```python
np.dot(sinx,cosx)
```

```python
np.outer(sinx,cosx)
```
