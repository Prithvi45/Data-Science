## Difference Between loc, iloc and ix
* loc needs index to find data - loc[1]  ->it will fetch index=1 data , loc[2] ->index=2 data, loc[:2] -index=0,1,2
* iloc : this uses slicing method iloc[:2] -> index values upto 2 >> 0 and 1
* ix : similar to loc  but e.g df.iloc[:2,:3] -> first it takes index 0,1,2 and columns as slicing 0,1,2.


```python
goal_data.loc[:2]
```

```python
goal_data.ix[:2]
```

```python
goal_data.iloc[:2]
```

```python
dA = goal_data.ix[:2,:3]
dA
```

```python
dB = goal_data[2:]
dB
```

```python
dA.append(dB)
```
