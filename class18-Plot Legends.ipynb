```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,1000)
```

```python
#fig - contains all the subplots
#ax - array of all subplots
fig,ax = plt.subplots()
ax.plot(x,np.cos(x),label='-sin')
ax.plot(x,np.sin(x), label='-cos')
leg =ax.legend()
```
