```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,1000)
```

```python
#fig - contains all the subplots
#ax - array of all subplots
fig,ax = plt.subplots()
ax.plot(x,np.cos(x),label='-sin')
ax.plot(x,np.sin(x), label='-cos')
leg =ax.legend()
```

```python

import pandas as pd
cities = pd.read_csv('california_cities.csv')
```

```python
cities.head()
```

```python
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python

plt.scatter(lon, lat, c = np.log10(population), s=area, alpha=0.5 )
plt.colorbar(label='log population')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('Population Density')
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.4, wspace=0.4)
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5, str(i))
```

```python
births = pd.read_csv('births.csv')
```

```python
births.head()
```

```python
births['day'] = births['day'].astype(int)

births.index = pd.to_datetime(10000 * births.year +
                              100 * births.month +
                              births.day, format='%Y%m%d')
births_by_date = births.pivot_table('births',
                                     [births.index.month, births.index.day])
births_by_date.index = [pd.datetime(2012, month, day)
                         for (month, day) in births_by_date.index]
```
